Now that we have a filtered dataset with the artist-number_of_apparition association for each years. The idea is now to take the 5 most present rappers for each decades

```python
import os
#Open the files
import_dir = "./Result_filtered"
j = 1
decades = {}
general_dict = {}
decade_nb = 0
for filename in os.listdir(import_dir):
    most_famous_rappers = {}
    #Import each files
    with open(os.path.join(import_dir, filename), 'r') as file:
        # Iterate over each line in the file
        i = 1
        for line in file:
            # Split the line using the separator "-"
            result = line.rsplit("-", 1)
            # Add the key-value pair to the dictionary
            most_famous_rappers[result[0]] = int(result[1].strip())
    
    for key, value in most_famous_rappers.items():
        if key in general_dict:
            general_dict[key] += value
        else:
            general_dict[key] = value
    if j%10 == 0:
        decades[decade_nb] = sorted(general_dict, key=general_dict.get, reverse=True)[:100]
        general_dict = {}
        decade_nb += 1
    j+=1
    most_famous_rappers = {}
    # Sort the dictionary by value
    # print(f"Top 10 rappers in {filename[29:38]}")
    # print(most_famous_rappers)
print(decades)
```

we have now a list of influent rappers for the 1990's, 2000's and 2010's decades.

```python
all_rappers = []
for key, value in decades.items():
   # all_rappers += [ x for x in value if x not in all_rappers]
   all_rappers += value
all_rappers
```
